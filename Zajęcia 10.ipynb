###### Matlibplot
    Jest to biblioteka do wizualizacji danych (tworzenia wykresów). Ważniejsze funkcjonalności znajdują 
    się w module matlibplot.pyplot. Biblioteka wykorzystuje tablice z NumPy, automatycznie konwertując 
    inne typy danych do ndarray.
    
    Funkcja plot() - wykreśla podane jej punkty. Podanie jednej serii spowoduje wyświetlenie ich względem 
    osi x; Przekazanie funkcji plot() dwóch serii punktów pozwoli na wykreślenie zależności x(y).
    Funkcja show() - wyświetla stworzone wykresy.
    
    Formatowanie wykresów:
        * Trzecim argumentem funkcji plot() jest styl serii danych. Podawane "[rodzaj pkt][rodzaj linii][kolor]"
            > rodzaje pkt '.', ',', 'o', '^', 's', '*'
            > rodzaje linii '-', '--', '-.', ':'
            > kolory 'b', 'g', 'r', 'm' 
        * Za zakres wartości na osiach odpowiada funkcja plt.axis(xmin,xmax,ymin,ymax)
        * Za opis odpowiadają funkcje plt.xlabel() i plt.ylabel()
        * Legenda może być dodana funkcją plt.legend([lista podpisów])
        * Tytuł wykresu można dodać funkcją plt.title()

    Kilka wykresów:
        Aby stworzyć kilka wykresów obok siebie należy użyć funkcji plt.subplot(), tworzącej nowy wykres, 
        do którego będą się odwoływać kolejne komendy. Funkcja ta przyjmuje trzy inty oznaczające kolejno: 
        ilość rzędów, ilość kolumn i indeks wykresu.
    

```python
# Tworzenie wykresu
import matplotlib.pyplot as plt

plt.plot([1,5,10,15]) # tworzy wykres z następującymi punktami
plt.show() # wyświetlenie wykresu

xpt = [1,2,3,4] # tworzenie punktów
ypt = [0,5,10,15]
plt.plot(xpt,ypt) # wykres dwóch serii punktów
plt.show()
```

```python
# Formatowanie wykresów
import matplotlib.pyplot as plt

xpt1 = [1,2,3,4]
ypt1 = [0,5,10,15]
xpt2 = [1,2,3,4]
ypt2 = [0,2,4,6]

plt.plot(xpt1,ypt1,'-b',xpt2,ypt2,'--g') # zmienne i format
plt.axis([1,4,0,15]) # zakres osi
plt.xlabel("Oś x") # podpisy osi
plt.ylabel("Oś y")
plt.title("Tytuł wykresu") # tytuł wykresu
plt.show()
```

```python
# Kilka wykresów
import matplotlib.pyplot as plt

plt.subplot(121) # Pierwszy podwykres
xpoints1 = [1,2,3,4]
ypoints1 = [2,3,5,6]
plt.plot(xpoints1, ypoints1, '-r')
plt.axis([1,4,2,10])
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Tytuł1')

plt.subplot(122) # Drugi podwykres
xpoints2 = [1,2,3,4]
ypoints2 = [4,5,6,7]
plt.plot(xpoints2, ypoints2, '--g')
plt.axis([1,4,2,10])
plt.xlabel('oś X')
plt.title('Tytuł2')

plt.show() # Pokazanie obu wykresów
```

I. Napisz program wczytujący dwa pliki zawiarające wartości funkcji rozkładu radialnego (RDF_O1H.xvg i RDF_O1H.xvg) i wyświetlający ich nakładające się wykresy. 

```python
# import pakietów
import numpy as np
import matplotlib.pyplot as plt

# wykrajanie danych
a=[]
b=[]
c=[]
d=[]
with open("RDF_O1H.xvg") as f1: # otwarcie pliku ...O1H.xvg
    for linia in f1: # pocięcie pliku
        t = linia.split()
        if linia[0] == "#": # pominięcie komentarzy w pliku
            pass
        elif linia[0] == "@":
            pass
        else: # dodanie wartości do list a i b
            a.append(float(t[0]))
            b.append(float(t[1]))
with open("RDF_O2H.xvg") as f2: # otwarcie pliku ...O2H.xvg
    for linia in f2: # pocięcie pliku
        y = linia.split()
        if linia[0] == "#": # pominięcie komentarzy w pliku
            pass
        elif linia[0] == "@":
            pass
        else: # dodanie wartości do list c i d
            c.append(float(y[0]))
            d.append(float(y[1]))

# edycja wykresu
plt.plot(a,b,'-r',c,d,'-g')
plt.legend(['RDF_O1H','RDF_O2H'])
plt.xlabel('r[nm]')
plt.ylabel('ro[r]')
plt.title('Porównanie rozkładu radialnego')

# wyrysowanie wykresu
plt.show()
```

II. Napisz program rysujący funkcje sinusa i cosinusa (w rozsądnym przedziale) i wyznaczający na wykresie punkty przecinania się funkcji (z rozsądną tolerancją).

```python
# import pakietów
import numpy as np
import matplotlib.pyplot as plt

# tworzenie tablic i list
a = np.arange(0, 2*np.pi, 0.01*np.pi)
y1 = []
y2 = []
pkt1 = []

# pętle
for i in a: # zapełnienie list wartościami sinusa i cosinusa
    b = np.sin(i)
    c = np.cos(i)
    y1.append(b)
    y2.append(c)
for h in range(len(a)): # tworzenie punktów przecinających się 
    if abs(y2[h]-y1[h]) < 0.01: # kryterium zbieżności
        pkt1.append(y2[h])
    else:
        pkt1.append(np.nan)

# edycja i wypisanie wykresu
plt.plot(a,y1,"-r",a,y2,"-b",a,pkt1,"o-g")
plt.show()
```
