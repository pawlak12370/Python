```python
import numpy as np
from scipy.integrate import quad

def f(x):
    return np.sin(x*np.pi)

res, err = quad(f, 0, 1)
print(res, err)
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(y,t):
    dydt = 2*y
    return dydt

t = np.linspace(0,4,200)
y0 = 0.001
solv = odeint(f, y0, t)

plt.plot(t, solv, '--r')
plt.show()
```

```python
# Import modułów
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
import scipy.constants as scpc

# Zadanie zmiennych
a = int(input("Wpisz granicę pudła: "))
n = int(input("Wpisz poziom energetyczny: "))
m = int(input("Wpisz masę molekuły: "))

#Tworzenie funkcji
def f(x):
    return ((2/a)**0.5)*np.sin(((n*np.pi)/a)*x)

def fkw(x):
    return (((2/a)**0.5)*np.sin(((n*np.pi)/a)*x))**2

#Tworzenie wykresu
x = np.linspace(0,a,100)
plt.plot(x,f(x),"-r")
plt.show()

#Unormowanie funkcji
res = quad(fkw,0,a)
if res == 1:
    print("Funkcja jest unormowana")
else:
    print("Funkcja nie jest unormowana")
    
#Energia 
energia = ((n**2)*(scpc.Planck**2))/(8*m*a**2)
print("Energia wynosi: ",energia)

print()
print()
print(" _   |~  _")
print("[_]--'--[_]")
print("|'|--|--|'|")
print("| | /^\ | |")
print("|_|_|I|_|_|")
```

```python
# Import modułów
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import scipy.constants as scpc

# Zadanie zmiennych
a = float(input("Podaj początkowe stężenie substratu A: "))
b = float(input("Podaj początkowe stężenie substratu B: "))
k = float(input("Podaj stałą szybkości reakcji: "))

# Definicja funkcji
def f(x,t,a,b,k):
    dxdt = k*(a-x)*(b-x)
    return dxdt

# Budowanie wykresu
t = np.linspace(0,10,100)
x0 = 0
solv = odeint(f,x0,t,args=(a,b,k))

plt.plot(t,solv,'-b',t,a-solv,'-r',t,b-solv,'-g')
plt.legend(['produkt C','substrat A','substrat B'])
plt.xlabel('Czas')
plt.ylabel('Stężenie')
plt.title('Kinetyka Reakcji')
plt.cool()
plt.show()

print()
print()
print(" _   |~  _")
print("[_]--'--[_]")
print("|'|--|--|'|")
print("| | /^\ | |")
print("|_|_|I|_|_|")
```
